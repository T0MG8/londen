```python
from folium import plugins

link_station1 = r"C:\Users\tomgo\OneDrive\Bureaublad\Data Science\Case 3 - Londen\Data\Londen data\London stations.JSON"
with open(link_station1, "r", encoding="utf-8") as a:
    data = json.load(a)

# Extract features en zet ze om naar een DataFrame
features = data["features"]
station1 = pd.json_normalize(features, sep="_")

# Kleuren per zone (zelfde als eerder)
zone_colors = {
    "1": "red",
    "2": "orange",
    "3": "yellow",
    "4": "lime",
    "5": "cyan",
    "6": "blue"
}

button_1 = [{'label': "Zones", 'method': "update"}]
button_2 = [{'label': "Drukte", 'method': "update"}]


m = folium.Map(location=[51.5, -0.1], zoom_start=10)

plugins.Draw().add_to(m)

for idx, row in station1.iterrows(): 
    coords = row['geometry_coordinates'] 
    
    lat = coords[1] 
    lon = coords[0] 
    
    zone = row.get("properties_zone", "6") 
    color = zone_colors.get(zone, "gray") 
    
    folium.CircleMarker(
        location=[lat, lon], 
        radius=8,
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7,
        popup=row["properties_name"]  
    ).add_to(m)

m
```

```python
df_list = [bez_2007, bez_2008, bez_2009, bez_2010, bez_2011, bez_2012, bez_2013, bez_2014, bez_2015, bez_2016, bez_2017]
years = range(2007, 2018)

# Verzamel Entry en Exit per jaar
df_entry_exit = pd.concat(
    [df.set_index(["nlc", "Station"])[["Entry_Week", "Exit_Week"]].rename(
        columns={"Entry_Week": f"Entry_Week_{year}", "Exit_Week": f"Exit_Week_{year}"}
    ) for df, year in zip(df_list, years)],
    axis=1
)

# Voeg properties_zone toe aan entry_exit_week door te mergen op de stationnaam
entry_exit_week = entry_exit_week.merge(
    station1[["properties_name", "properties_zone"]],
    left_on="Station", right_on="properties_name",
    how="left"
)

# Kolom hernoemen naar 'Zone' en de overbodige 'properties_name' verwijderen
entry_exit_week = entry_exit_week.rename(columns={"properties_zone": "Zone"}).drop(columns=["properties_name"])

# Bekijk de eerste rijen
entry_exit_week.head()
```

```python
import plotly.express as px

# Tel het aantal NaN's per kolom
nan_counts = entry_exit_week.isna().sum()

# Zet de data in een dataframe
nan_df = nan_counts.reset_index()
nan_df.columns = ["Kolom", "Aantal_NaN"]

# Maak een staafdiagram met Plotly
fig = px.bar(nan_df, x="Kolom", y="Aantal_NaN", title="Aantal NaN's per kolom", labels={"Aantal_NaN": "Aantal NaN's"})

# Toon de plot
fig.show()
```
